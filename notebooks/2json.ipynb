```python
import numpy as np
import pandas as pd
import os
from countrycode import countrycode
import json
```

```python
dates = []

for file in os.listdir("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts"):
    dates.append(file[:8])

with open("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/dates.json", "w") as f:
    json.dump(dates, f)
```

```python
gkg = None
for file in os.listdir("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts"):
    temp = pd.read_csv(f"C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts/{file}", header=0, delimiter="\t")
    if gkg is not None:
        gkg = pd.concat([gkg, temp], ignore_index=True)

    else:
        gkg = temp


gkg.dropna(subset=["GEO_COUNTRYCODE"], inplace=True)
```

```python
new = gkg.groupby("GEO_COUNTRYCODE").agg(count=("NUMARTS", sum))#, urls=("SOURCEURLS", list))
new["iso3"] = countrycode(new.index, origin='fips', destination='iso3c')
new.dropna(subset=["iso3"], inplace=True)
json_data = new.to_dict(orient="records")


with open("counts.json", "w") as f:
    json.dump(json_data, f)
```
