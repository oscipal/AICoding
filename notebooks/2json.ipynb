```python
import numpy as np
import pandas as pd
import os
from countrycode import countrycode
import json
```

```python
dates = []

for file in os.listdir("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts"):
    dates.append(file[:8])

with open("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/dates.json", "w") as f:
    json.dump(dates, f)
```

```python
for file in os.listdir("C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts"):

    gkg = pd.read_csv(f"C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/data/gkgcounts/{file}", header=0, delimiter="\t")

    gkg.dropna(subset=["GEO_COUNTRYCODE"], inplace=True)

    new = gkg.groupby("GEO_COUNTRYCODE").agg(count=("NUMARTS", sum))

    new["iso3"] = countrycode(new.index, origin='fips', destination='iso3c')

    new.dropna(subset=["iso3"], inplace=True)
    json_data = new.to_dict(orient="records")

    with open(f"C:/Users/ottos/Documents/Geomatik/2. Sem/AICoding/mapbox/{file[:8]}.json", "w") as f:
        json.dump(json_data, f)
```
